```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("D:\MCA-4th semester\data science dataset\IPL Ball-by-Ball 2008-2020.csv")
data.head(10)
```

```python
data.info()
```

```python
temp = pd.DataFrame({"total_runs": data['total_runs']})
count_runs = temp.value_counts()
print(count_runs)
labels = [X[0] for X in count_runs.keys()]


bar, ax = plt.subplots(figsize=(20,12))
ax = plt.pie(x=  count_runs,autopct="%.1f%%",labels=labels)
plt.title("most runs in IPL",fontsize=15)
plt.show()
```
